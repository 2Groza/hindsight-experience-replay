```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.pyplot import MultipleLocator

sns.set_style('white')
for key in result.keys():
    results_now = np.concatenate((np.asarray(result[key][1]).reshape(1,-1), np.asarray(result[key][2]).reshape(1,-1), np.asarray(result[key][3]).reshape(1,-1), np.asarray(result[key][4]).reshape(1,-1)), 0)
    plt.plot(results_now.mean(0), label = key)
plt.title('FetchReach-v1', fontsize=15)
plt.xlabel('Epoch', fontsize = 15)
plt.ylabel('Success Rate', fontsize = 15)
plt.legend(loc='lower right')
x_major_locator=MultipleLocator(5)
y_major_locator=MultipleLocator(0.2)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)

plt.xlim(-0.5,20.5)

plt.show()
    
```
